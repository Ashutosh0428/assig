```python
#4 . write a fun which will take another function as an input and return me an output 
def add(a, b):
    return a + b

def mult(a, b):
    return a * b

def add_and_mult(a, b):
    added = add(a, b)
    multed = mult(a, b)
    return add(added, multed)
```

```python
add_and_mult(40, 50)
```

```python
#5 write a function which will take multiple list as a input and give me concatnation of all the element as and output 
def concat(list):
    result= ''
    for element in list:
        result += str(element)
    return result
```

```python
concat([1,2,3,'ashu',45,])
```

```python
print(concat([1,23,43,43]))
```
